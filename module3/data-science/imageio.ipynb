# imageio

Python library for reading and writing image data

```
from imageio import imread, imsave
import numpy as np
```

# Read an image as a NumPy array

```
url = 'https://opengameart.org/sites/default/files/Preview%203_0.png'
img = imread(url)
```

# How to get a grayscale image from the raw one

```
def rgb2gray(rgb):
  return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
```

```
grayscale = rgb2gray(img)
imsave('/content/grayscale.jpg', grayscale)
```

# How to get the negative of the previous image from its grayscale

```
def negative(bw, threshold=255/2):
  bw[bw < threshold] = 0    # Black
  bw[bw >= threshold] = 255 # White

  return bw
```

```
# Black and white
bw = negative(grayscale, threshold=50)
imsave('/content/negative.jpg', bw)
```

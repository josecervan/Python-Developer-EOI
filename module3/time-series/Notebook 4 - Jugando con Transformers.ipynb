*Hugging Face* es una librería que nos permite trabajar con una gran variedad de modelos basados en Transformers de una manera sencilla.

El modo más simple de interactuar con los modelos de esta librería es utilizando el método ***pipeline()***. Esta función se encargará automáticamente de tokenizar tu texto y de hacer uso del modelo adecuado para resolver tu tarea. Por ejemplo, si quisieramos utilizar un clasificador de sentimientos, esto sería tan sencillo como:



```
from transformers import pipeline
```

```
# Creamos un clasificador para análisis de sentimiento.
classifier = pipeline('sentiment-analysis')
# Y probamos qué tan bien funciona.
print(classifier('We are very happy to show you the Transformers library.'))
print(classifier('I hope this library is not as bad as programming everything from scratch.'))
```

```
# Creamos un clasificador para análisis de sentimiento con un modelo personalizado.
classifier = pipeline('sentiment-analysis', model='nlptown/bert-base-multilingual-uncased-sentiment')
# Probamos nuévamente el modelo en Español.
print(classifier('¡Está super bien esta librería!'))
print(classifier('Espero que esto no sea tan pesado como tener que programar todo desde cero.'))
print(classifier('Porque detestaría tener que hacerlo.'))
```

- **Carga el modelo T5 en tu código**, junto a su tokenizador para transformar el input y generar un resultado con el modelo. Para esto te interesará buscar en la documentación información sobre *T5Tokenizer* y* TFT5ForConditionalGeneration*

```
!pip install transformers
```

```
from transformers import T5Tokenizer, T5ForConditionalGeneration

model_name = "t5-base"

tokenizer = T5Tokenizer.from_pretrained(model_name)
```

```
inputs = tokenizer("summarize: state authorities dispatched emergency crews tuesday to survey the damage after an on slaught of severe weather in mississippi…", return_tensors='pt').input_ids

model = T5ForConditionalGeneration.from_pretrained(model_name, return_dict=True)

outputs = model.generate(inputs)

decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(decoded)
```

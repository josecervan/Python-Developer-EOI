```
#Eliminar emojis
original_tweet="RT MARAVILLOSO ğŸ‘ğŸ‘ Siempre recordaremos la voz y gen diferencial que tuvo Michael para comentar los partidos @InformeRobinson #Great #Person https://t.co/XVEXKQq9Yq"
texto = original_tweet.encode("ascii","ignore")
texto = str(texto, 'utf-8')
texto
```

```
#Eliminar urls y RT

texto="RT MARAVILLOSO ğŸ‘ğŸ‘ Siempre recordaremos la voz y gen diferencial que tuvo Michael para comentar los partidos @InformeRobinson #Great #Person https://t.co/XVEXKQq9Yq"
tweet = re.sub('http\S+\s*', '', texto)  # remove URLs
tweet = re.sub('RT|cc', '', tweet)
#texto
tweet
```

```
#FunciÃ³n de limpieza: asÃ­ es como debemos trabajar
#https://www.machinelearningplus.com/python/python-regex-tutorial-examples/

def clean_tweet(tweet):
    tweet = re.sub('http\S+\s*', '', tweet)  # remove URLs
    tweet = re.sub('RT|cc', '', tweet)  # remove RT and cc
    tweet = re.sub('#\S+', '', tweet)  # remove hashtags
    tweet = re.sub('@\S+', '', tweet)  # remove mentions
    tweet = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), '', tweet)  # remove punctuations
    tweet = re.sub('\s+', ' ', tweet)  # remove extra whitespace
    return tweet

texto = "RT MARAVILLOSO ğŸ‘ğŸ‘ Siempre recordaremos la voz y gen diferencial que tuvo Michael para comentar los partidos @InformeRobinson #Great #Person https://t.co/XVEXKQq9Yq"

print(clean_tweet(texto))
```

```
#Paso a minusculas
texto = texto.lower()
texto
```

## Ejercicios WordCount

```
#Definimos la funcion para contar palabras 
def word_count(str):
    counts = dict()
    words = str.split()
 
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
 
    return counts
```

```
res = word_count("mi prueba es una prueba buena")
res
```

```
#Ordenar el resultado
type(res)
orden = sorted(res.items(), key=lambda x: x[1])
orden
```

```
import pandas as pd
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
#nltk.download()  #Solo hacerlo una vez!!!
nltk.download('stopwords')

data = pd.read_csv("data_clean.csv") 
pd.set_option('display.max_columns', None)  
data.head(20)
```

```
#Convertimos toda la columna Tweet en un string

text = " ".join(tweet for tweet in data.Tweet)


# Generamos la wordcloud
wordcloud = WordCloud(max_font_size=50, max_words=80, background_color="white").generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
#Repetimos la operaciÃ³n pero esta vez aplicando el filtro de stopwords

stop_words = set(stopwords.words('english')) 
nltk.download('punkt')

# Generate a word cloud image
word_tokens = word_tokenize(text)
filtered_tweets = [w for w in word_tokens if not w in stop_words]
print(filtered_tweets)
clean_text = " ".join(filtered_tweets)
wordcloud = WordCloud(max_font_size=50, max_words=80, background_color="white").generate(clean_text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
